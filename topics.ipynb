```python
def process_node(level,node):
    lines=[]
    sep="\t"
    prefix=sep*level
    sufcount = 3-level
    suffix=sep*sufcount
    defin = g.value(node,def_prop)
    if defin == None:
        defin ="**No Definition***"
    entry = prefix+g.label(node)+sep+suffix+"\""+defin+"\"\n"
    print(entry)
    lines.append(entry)
    for s in g.subjects(RDFS.subClassOf,node):
        sublines = process_node(level+1,s)
        lines = lines + sublines
    return lines
```

```python
lines = process_node(0,topic)
```
